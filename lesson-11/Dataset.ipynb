```scala
def computeTemp(temp: Long): Double = {
    if (temp > 25) temp * 0.15 else temp * 0.5
}
```

```scala
ds.map(d => computeTemp(d.temp)).show(5, false)
```

### Проекция

```scala
ds
    .filter(d => {d.temp > 25})
    .map(d => (d.temp, d.device_name, d.device_id, d.cca3))
```

```scala
ds
    .filter(d => {d.temp > 25}).map(d => (d.temp, d.device_name, d.device_id, d.cca3))
    .withColumnRenamed("_1", "temp")
    .withColumnRenamed("_2", "device_name")
    .withColumnRenamed("_3", "device_id")
    .withColumnRenamed("_4", "cca3")
```

```scala
ds
    .filter(d => {d.temp > 25})
    .map(d => (d.temp, d.device_name, d.device_id, d.cca3))
    .toDF("temp", "device_name", "device_id", "cca3")
```

```scala
case class DeviceTempByCountry(
    temp: Long,
    device_name: String,
    device_id: Long,
    cca3: String
)
```

```scala
val dsTemp = ds
    .filter(d => {d.temp > 25})
    .map(d => (d.temp, d.device_name, d.device_id, d.cca3))
    .toDF("temp", "device_name", "device_id", "cca3")
    .as[DeviceTempByCountry]
```

```scala
dsTemp.show(5, false)
```

```scala
ds
    .select($"battery_level", $"c02_level", $"device_name")
    .where($"battery_level" < 8)
    .sort($"c02_level")
    .show(5, false)
```

```scala
ds.select(
    min("temp"),
    max("temp"),
    min("humidity"),
    max("humidity"),
    min("c02_level"),
    max("c02_level"),
    min("battery_level"),
    max("battery_level")
).show()
```

### groupBy и агрегации

```scala
val newDS = ds
    .filter(d => {d.c02_level > 1300})
    .groupBy($"cn")
    .avg()
    .sort($"avg(c02_level)".desc)

newDS.show(5, false)
```

```scala
ds
    .filter(d => {d.temp > 25 && d.humidity > 75})
    .select("temp", "humidity", "cn")
    .groupBy($"cn")
    .avg()
    .sort($"avg(temp)".desc, $"avg(humidity)".desc)
    .show(5, false)
```

### groupByKey и агрегации

```scala
val populationByCity: Dataset[(String, Long)] =
	people.groupByKey(_.city)
	.agg(count(lit(1)).as[Long])

populationByCity.show
```

```scala
val cityStats = people.groupByKey(_.city)
	.mapGroups { case (city, iter) => (city, iter.map(_.age).sum) }

cityStats.show
```

### reduceGroups

```scala
val oldestByCity = people
	.groupByKey(_.city)
	.reduceGroups((p1, p2) => if (p1.age > p2.age) p1 else p2)
	.map(_._2)

oldestByCity.show
```

### joinWith

```scala
case class City(name: String, country: String)
```

```scala
val cities = Seq(
    City("Moscow", "Russia"),
    City("SPb", "Russia"),
    City("Ekaterinburg", "Russia"),
    City("Novosibirsk", "Russia"),
    City("Krasnodar", "Russia")
).toDS()
```

```scala
val ij = people.joinWith(cities, people("city") === cities("name"), "inner")
ij.show(false)
```

```scala
val lj = people.joinWith(cities, people("city") === cities("name"), "left")
lj.show(false)
```
